```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
store_df= pd.read_csv("store.csv")
store_df.head()
```

```python
stores_df= store_df.sort_values(by=['Profit'],ascending= False).reset_index(drop=True)
stores_df.tail(10)
```

```python
stores_df.groupby("Ship Mode").nunique()
```

```python
cat_df= stores_df.groupby(["Category","Sub-Category"])["Profit"].sum().sort_values(ascending=False).reset_index()
cat_df.head(50)
```

```python
cat_df.shape
```

```python
fig= plt.figure(figsize=(20,15))
plt.bar(cat_df['Sub-Category'],cat_df['Profit'],color="red")
plt.xlabel("Categories")
plt.ylabel("Profits")
plt.show()
```

```python
region_df= stores_df.groupby("Region")['Profit'].sum().sort_values(ascending=False).reset_index()
region_df.head()
```

```python
plt.bar(region_df['Region'],region_df['Profit'])
```

```python
stores_df.head()
```

```python
regions_df= stores_df.groupby("Region")["Sub-Category"].value_counts()
regions_df.head(52)
```
