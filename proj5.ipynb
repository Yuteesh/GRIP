```python
deliveries_df= deliveries_df.fillna(0)
deliveries_df.head(50)
```

```python
deliveries_df['dismissal_kind']= deliveries_df['dismissal_kind'].replace(to_replace=['caught', 'bowled', 'run out', 'lbw', 'caught and bowled','stumped', 'retired hurt', 'hit wicket', 'obstructing the field'],value=1)

deliveries_df.head(50)
```

```python
bat_df= pd.DataFrame()
```

```python
bat_df=deliveries_df.groupby(["batsman","batting_team"])['total_runs'].sum().nlargest(20).reset_index()
```

```python
bat_df.columns=['Batsman','Batting Team','Runs']
bat_df.index+=1
```

```python
bat_df
```

```python
bowl_df= pd.DataFrame()
```

```python
bowl_df= deliveries_df.groupby(['bowler','bowling_team'])['dismissal_kind'].sum().nlargest(20).reset_index()
bowl_df.columns=['Bowler','Team','No.of.Wickets']
bowl_df.index+=1
```

```python
bowl_df.head(50)
```

```python
player_df= pd.DataFrame()
```

```python
player_df= matches_df[['player_of_match','winner']].value_counts().nlargest(20).reset_index()
player_df.columns=['Man of Match','Team','No.of Times']
player_df.index+=1
```

```python
player_df.head(20)
```

```python
fig= plt.figure(figsize=(15,10))
plt.bar(toss_df['Winner'],toss_df['Count'])
plt.title("Toss based winner")
plt.xlabel("Winner")
plt.ylabel("Number of times")
plt.show()
```
